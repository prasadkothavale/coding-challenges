```python
count = 0
for i, row in tqdm(enumerate(grid)):
  for j, cell in enumerate(row):
    adj = 0

    if i > 0 and j > 0 and grid[i-1][j-1]:
      adj += 1

    if i > 0 and grid[i-1][j]:
      adj += 1

    if i > 0 and j < len(row)-1 and grid[i-1][j+1]:
      adj += 1

    if j > 0 and grid[i][j-1]:
      adj += 1

    if j < len(row)-1 and grid[i][j+1]:
      adj += 1

    if i < len(row)-1 and j > 0 and grid[i+1][j-1]:
      adj += 1

    if i < len(row)-1 and grid[i+1][j]:
      adj += 1

    if i < len(row)-1 and j < len(row)-1 and grid[i+1][j+1]:
      adj += 1

    if cell and adj < 4:
      count += 1

print('--------')
print(count)
print('--------')
```

## Part Two

```python
def analyse_and_update_grid(grid):
  grid = copy.deepcopy(grid)
  count = 0
  for i, row in enumerate(grid):
    for j, cell in enumerate(row):
      adj = 0

      if i > 0 and j > 0 and grid[i-1][j-1]:
        adj += 1

      if i > 0 and grid[i-1][j]:
        adj += 1

      if i > 0 and j < len(row)-1 and grid[i-1][j+1]:
        adj += 1

      if j > 0 and grid[i][j-1]:
        adj += 1

      if j < len(row)-1 and grid[i][j+1]:
        adj += 1

      if i < len(row)-1 and j > 0 and grid[i+1][j-1]:
        adj += 1

      if i < len(row)-1 and grid[i+1][j]:
        adj += 1

      if i < len(row)-1 and j < len(row)-1 and grid[i+1][j+1]:
        adj += 1

      if cell and adj < 4:
        count += 1
        grid[i][j] = False

  #print(count)
  return [count, grid]


count = 0
prev_count = -1
new_grid = copy.deepcopy(grid)

while count > prev_count:
  prev_count = count
  [new_count, new_grid] = analyse_and_update_grid(new_grid)
  count += new_count


print('--------')
print(count)
print('--------')
```
