```python
class Junction:
  def __init__(self, x, y, z, index):
    self.x = x
    self.y = y
    self.z = z
    self.index = index

  def __repr__(self):
    return f'{self.index}: [{self.x}, {self.y}, {self.z}]'

class JunctionPair:
  def __init__(self, a, b, distance):
    self.a = a
    self.b = b
    self.distance = distance

  def __repr__(self):
    return f'{self.a.index} - {self.b.index}: {self.distance}'
```
