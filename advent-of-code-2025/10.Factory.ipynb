```python
import re

class Sample:
    def __init__(self, lights, buttons, voltage):
        self.lights = lights          # list[bool]
        self.buttons = buttons        # list[list[int]]
        self.voltage = voltage        # list[int]

    def __repr__(self):
        return f"Sample(lights={self.lights}, buttons={self.buttons}, voltage={self.voltage})"


def parse_samples(text):
    samples = []
    lines = [ln.strip() for ln in text.strip().split('\n') if ln.strip()]

    for line in lines:
        # 1. Lights: extract [.....]
        lights_str = re.search(r'\[([.#]+)\]', line).group(1)
        lights = [c == '#' for c in lights_str]

        # 2. Buttons: extract all (...) groups
        button_groups = re.findall(r'\((.*?)\)', line)
        buttons = []
        for bg in button_groups:
            if bg.strip() == "":
                buttons.append([])     # empty tuple
            else:
                buttons.append(list(map(int, bg.split(','))))

        # 3. Voltage: extract {..}
        voltage_str = re.search(r'\{(.*?)\}', line).group(1)
        voltage = list(map(int, voltage_str.split(',')))

        samples.append(Sample(lights, buttons, voltage))

    return samples
```

```python
from collections import deque

class Sample:
    def __init__(self, lights, buttons, voltage):
        self.lights = lights          # final target state (boolean list)
        self.buttons = buttons        # list[list[int]]

    def bfs_depth(self):
        target = tuple(self.lights)
        n = len(target)

        start = tuple([False] * n)

        # If target is all-zero, depth = 0
        if start == target:
            return 0

        queue = deque([(start, 0)])   # (state, depth)
        visited = {start}

        while queue:
            state, depth = queue.popleft()
            next_depth = depth + 1

            for btn in self.buttons:
                new_state = list(state)

                # toggle lights for this button
                for pos in btn:
                    new_state[pos] = not new_state[pos]

                new_state = tuple(new_state)

                if new_state == target:
                    return next_depth

                if new_state not in visited:
                    visited.add(new_state)
                    queue.append((new_state, next_depth))

        return None  # theoretically unreachable if solution always exists


# -------------------------------------------------------------
# Parse function from previous message
# -------------------------------------------------------------

import re

def parse_samples(text):
    samples = []
    lines = [ln.strip() for ln in text.strip().split('\n') if ln.strip()]

    for line in lines:
        lights_str = re.search(r'\[([.#]+)\]', line).group(1)
        lights = [c == '#' for c in lights_str]

        button_groups = re.findall(r'\((.*?)\)', line)
        buttons = []
        for bg in button_groups:
            if bg.strip() == "":
                buttons.append([])
            else:
                buttons.append(list(map(int, bg.split(','))))

        # voltage exists but we no longer use it
        samples.append(Sample(lights, buttons, voltage=None))

    return samples


# -------------------------------------------------------------
# Example usage
# -------------------------------------------------------------

samples = parse_samples(test)

total_depth = 0
for i, s in enumerate(samples):
    d = s.bfs_depth()
    print(f"Sample {i+1} depth = {d}")
    total_depth += d

print("Total depth =", total_depth)
```

```python

from tqdm import tqdm
from functools import lru_cache

def parse_line(line):
    """
    Parse format:
    [.##.] (3) (1,3) (2) (2,3) (0,2) (0,1) {3,5,4,7}
    """
    parts = line.strip().split()

    # Ignore first token
    parts = parts[1:]

    # Last token = target
    target_token = parts[-1]
    target = list(map(int, target_token.strip("{}").split(",")))

    # Remaining tokens = buttons
    buttons = []
    for tok in parts[:-1]:
        tok = tok.strip("()")
        if tok == "":
            buttons.append(tuple())
        else:
            buttons.append(tuple(map(int, tok.split(","))))

    return buttons, target


def solve_one(buttons, target):
    N = len(target)

    # Build matrix A
    A = []
    for bx in buttons:
        row = [0] * N
        for c in bx:
            row[c] = 1
        A.append(tuple(row))

    # Compute upper bound for each button
    max_x = []
    for row in A:
        mx = float("inf")
        for i, v in enumerate(row):
            if v == 1:
                mx = min(mx, target[i])
        if mx == float("inf"):
            mx = 0
        max_x.append(mx)

    B = len(A)

    @lru_cache(None)
    def dfs(idx, t):
        t = list(t)

        if idx == B:
            return 0 if all(v == 0 for v in t) else None

        best = None
        row = A[idx]
        upper = max_x[idx]

        for x in range(upper + 1):
            new_t = t.copy()
            ok = True
            for i in range(N):
                if row[i] == 1:
                    new_t[i] -= x
                    if new_t[i] < 0:   # overshoot â†’ prune
                        ok = False
                        break

            if not ok:
                break  # increasing x will only get worse

            r = dfs(idx + 1, tuple(new_t))
            if r is not None:
                cand = r + x
                if best is None or cand < best:
                    best = cand

        return best

    return dfs(0, tuple(target))


def solve_all(text):
    lines = [ln.strip() for ln in text.strip().split("\n") if ln.strip()]
    results = []

    for ln in tqdm(lines, desc="Solving dataset"):
        buttons, target = parse_line(ln)
        ans = solve_one(buttons, target)
        results.append(ans)

    total = sum(r for r in results if r is not None)
    return results, total
```

```python

res, total = solve_all(test)
print("Per-line solutions:", res)
print("Sum =", total)
```
