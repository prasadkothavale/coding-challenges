- Data could take the connection from you to bbb, then from bbb to ddd, then from ddd to ggg, then from ggg to out.
- Data could take the connection to bbb, then to eee, then to out.
- Data could go to ccc, then ddd, then ggg, then out.
- Data could go to ccc, then eee, then out.
- Data could go to ccc, then fff, then out.

In total, there are 5 different paths leading from you to out.

How many different paths lead from you to out?

````python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
````

````python
def create_directional_graph_nx(data):
    G = nx.DiGraph()

    for current_node, next_nodes_list in data.items():
        for next_node in next_nodes_list:
            G.add_edge(current_node, next_node)

    return G
````

````python
input = test
input = input.split('\n')
data = {}
for line in input:
  [node, next] = line.strip().split(':')
  data[node] = next.strip().split(' ')

graph = create_directional_graph_nx(data)
nx.draw(graph, node_size = 1, arrowsize = 1)
#nx.draw_spring(graph, with_labels=True, node_color='skyblue', node_size=1500, edge_color='k', arrows=True, arrowsize=20)
#plt.show()
````

````python
all_paths_generator = nx.all_simple_paths(graph, source='you', target='out')
all_paths_list = list(all_paths_generator)
print(f'Total paths: {len(all_paths_list)}')
````

Thanks in part to your analysis, the Elves have figured out a little bit about the issue. They now know that the problematic data path passes through both dac (a digital-to-analog converter) and fft (a device which performs a fast Fourier transform).

```
svr,bbb,tty,ccc,eee,dac,fff,ggg,out
svr,bbb,tty,ccc,eee,dac,fff,hhh,out
```
However, only 2 paths from svr to out visit both dac and fft.

Find all of the paths that lead from svr to out. How many of those paths visit both dac and fft?

````python
def count_paths_fast(G, source, target):
    nx.set_node_attributes(G, 0, 'paths_to_out')

    G.nodes[target]['paths_to_out'] = 1
    topo_order = list(nx.topological_sort(G))

    for node in reversed(topo_order):
        if node == target:
            continue

        total_paths = 0
        for successor in G.successors(node):
            total_paths += G.nodes[successor]['paths_to_out']

        G.nodes[node]['paths_to_out'] = total_paths

    return G.nodes[source]['paths_to_out']
````

````python
total_count = count_paths_fast(graph, source='svr', target='out')
print(f"Total paths using DAG approach: {total_count}")
````

````python
def count_paths_via_targets(G, source, target, intermediate_nodes):
    node_dac, node_fft = intermediate_nodes

    # 1. Initialize four counters on every node
    # We use a dictionary for fast lookups: data[node] = { ... }
    # keys: 'total', 'has_dac', 'has_fft', 'has_both'
    memo = {node: {'total': 0, 'd': 0, 'f': 0, 'both': 0} for node in G.nodes()}

    # 2. Base Case: The 'out' node has 1 path to itself
    memo[target]['total'] = 1

    # 3. Get Reverse Topological Order (Children before Parents)
    try:
        # traverse reversed so we process end-nodes first
        reverse_order = reversed(list(nx.topological_sort(G)))
    except nx.NetworkXUnfeasible:
        return "Error: Graph has cycles"

    # 4. Traverse and Aggregate
    for u in reverse_order:
        if u == target:
            continue

        # --- A. Aggregate from children ---
        for v in G.successors(u):
            memo[u]['total'] += memo[v]['total']
            memo[u]['d']     += memo[v]['d']
            memo[u]['f']     += memo[v]['f']
            memo[u]['both']  += memo[v]['both']

        # --- B. Handle Logic for Special Nodes ---
        if u == node_dac:
            # All paths starting here now "have dac"
            memo[u]['d'] = memo[u]['total']
            # Any path that already had 'fft' now "has both"
            memo[u]['both'] = memo[u]['f']

        elif u == node_fft:
            # All paths starting here now "have fft"
            memo[u]['f'] = memo[u]['total']
            # Any path that already had 'dac' now "has both"
            memo[u]['both'] = memo[u]['d']

    # 5. Return the count of paths with 'both' starting from source
    return memo[source]['both']
````

````python
result = count_paths_via_targets(graph, 'svr', 'out', ('dac', 'fft'))
print(f"Paths visiting both 'dac' and 'fft': {result}")
````
